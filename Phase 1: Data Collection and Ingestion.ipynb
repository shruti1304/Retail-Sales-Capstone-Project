### Phase 1: Data Collection and Ingestion

#### Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

#### Import Dataset

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, when, isnan, isnull

# SparkSession is already created as 'spark', but this is good practice
spark = SparkSession.builder.appName("Retail_EDA").getOrCreate()

# Load datasets from DBFS
# inferSchema=True automatically detects numbers vs strings
df_sales = spark.read.csv("/Volumes/workspace/default/retail_sales/sales.csv/sales.csv", header=True, inferSchema=True)
df_stores = spark.read.csv("/Volumes/workspace/default/retail_sales/store_cities.csv", header=True, inferSchema=True)
df_products = spark.read.csv("/Volumes/workspace/default/retail_sales/product_hierarchy.csv", header=True, inferSchema=True)

print("Data loaded successfully!")
```

#### Display Schema and Row Counts

```python
print(f"Sales Data: {df_sales.count()} rows")
df_sales.printSchema()
```

```python
rows = df_sales.count()
cols = len(df_sales.columns)

print((rows, cols))
df_sales.display()
```

```python

print(f"\nStores Data: {df_stores.count()} rows")
df_stores.printSchema()
```

```python
rows = df_stores.count()
cols = len(df_stores.columns)

print((rows, cols))
df_stores.display()
```

```python
print(f"\nProducts Data: {df_products.count()} rows")
df_products.printSchema()
```

```python
rows = df_products.count()
cols = len(df_products.columns)

print((rows, cols))
df_products.display()
```

```python
df_sales.write.format("delta").mode("overwrite").saveAsTable("bronze_sales")
df_products.write.format("delta").mode("overwrite").saveAsTable("bronze_products")
df_stores.write.format("delta").mode("overwrite").saveAsTable("bronze_stores")
```
